```python
import pandas as pd
```

```python
code = pd.read_csv('country_code.csv')
code.head()
```

#### cancer : Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease

```python
cancer = pd.read_csv('30-70cancerChdEtc.csv')
cancer.head()
#merge
cancer_merge = code.merge(cancer, left_on='Country Name', right_on='Location', how = 'left')
cancer_merge = cancer_merge[cancer_merge['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
cancer_merge = cancer_merge.rename(columns = {'Period': 'year',
                                             'First Tooltip':'cancer'})
cancer_merge
```

#### air_pollution_death : Ambient and household air pollution attributable death rate (per 100 000 population, age-standardized)

```python
air_death = pd.read_csv('airPollutionDeathRate.csv')
air_death.head()
#merge
air_death_merge = code.merge(air_death, left_on='Country Name', right_on='Location', how = 'left')
air_death_merge = air_death_merge[air_death_merge['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
air_death_merge = air_death_merge.rename(columns = {'Period': 'year',
                                             'First Tooltip':'air_pollution_death'})
air_death_merge
```

#### alchol_abuse:Total (recorded+unrecorded) alcohol per capita (15+) consumption

```python
alchol_abuse = pd.read_csv('alcoholSubstanceAbuse.csv')
alchol_abuse.head()
#merge
alchol_abuse = code.merge(alchol_abuse, left_on='Country Name', right_on='Location', how = 'left')
alchol_abuse = alchol_abuse[alchol_abuse['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
alchol_abuse = alchol_abuse.rename(columns = {'Period': 'year',
                                             'First Tooltip':'alchol_abuse'})
alchol_abuse
```

#### sanitation: Population using at least basic sanitation services (%)

```python
sanitation = pd.read_csv('atLeastBasicSanitizationServices.csv')
sanitation.head()
#merge
sanitation = code.merge(sanitation, left_on='Country Name', right_on='Location', how = 'left')
sanitation = sanitation[sanitation['Dim1'] == 'Total'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
sanitation = sanitation.rename(columns = {'Period': 'year',
                                             'First Tooltip':'basic_sanitiz_service'})
sanitation
```

#### drinking_water: Population using at least basic drinking-water services (%)

```python
drinking_water = pd.read_csv('basicDrinkingWaterServices.csv')
drinking_water.head()
#merge
drinking_water = code.merge(drinking_water, left_on='Country Name', right_on='Location', how = 'left')
drinking_water = drinking_water[['Country Code', 'Country Name', 'Period', 'First Tooltip']]
drinking_water = drinking_water.rename(columns = {'Period': 'year',
                                             'First Tooltip':'drinking_wate'})
drinking_water
```
